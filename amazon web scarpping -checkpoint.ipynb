```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ="https://www.amazon.in/s?k=playstation+5&crid=3GKJOSMJK6N7M&sprefix=playstation+%2Caps%2C324&ref=nb_sb_noss_2"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(url, headers=HEADERS)
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
links = soup.find_all("a", attrs={'class': 'a-link-normal s-no-outline'})
```

```python
links
```

```python
links = links[0].get('href')
```

```python
product_list = "https://amazon.in" + links
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup =BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
new_soup.find("span", attrs={"id":'productTitle'}).text.strip()
```
